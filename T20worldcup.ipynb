```python
#identifying and handling missing value
df.replace('?', np.nan, inplace=True)
null_values = df.isnull()
null_values.sum()
```

```python
#dealing with missing value
wicket_new_df = df[["wickets_0_player_out","wickets_0_kind","wickets_0_fielders_0_name", "wickets_0_fielders_1_name"]]
wicket_new_df.fillna("NA", inplace = True)

extra_new_df = df[["extras_wides","extras_legbyes","extras_noballs", "extras_penalty"]]
extra_new_df.fillna(0, inplace = True)

review_new_df = df[["review_by","review_umpire","review_batter","review_decision", "review_type"]]
review_new_df.fillna("NA", inplace = True)
```

```python
#renaming columns names for better understanding
df.rename(columns = {"wickets_0_player_out":"player_out_name"}, inplace = "True")
df.rename(columns = {"wickets_0_kind":"wickets_type"}, inplace = "True")
df.rename(columns = {"wickets_0_fielders_0_name":"wicket_taken_by_fielder"}, inplace = "True")
df.rename(columns = {"wickets_0_fielders_1_name":"wicket_taken_by_fielder_2"}, inplace = "True")
```

```python
#Calculating total nums of runs
total_run_by_USA = 0
total_run_by_India = 0

for index, row in df.iterrows():
  if row["team"] == "United States of America":
    total_run_by_USA += row["runs_total"]
  else:
    total_run_by_India += row["runs_total"]

print("Total run by USA: ", total_run_by_USA)
print("Total run by India: ", total_run_by_India)

#Plotting total number of run graph

labels = ['USA', 'India']
values = [total_run_by_USA, total_run_by_India]
plt.bar(labels, values)
plt.xlabel("Teams")
plt.ylabel("Total Runs")
plt.title("Total Runs by each Team for IND Vs USA T20")
plt.show()
```

```python
#Calculating total nums of runs in each over
runs_by_over_team = df.groupby(['over', 'team'])['runs_total'].sum().unstack()
#print(runs_by_over_team)

#plotting bar plot
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = runs_by_over_team.index

bar1 = ax.bar(index - bar_width/2, runs_by_over_team['India'], bar_width, label='IND')
bar2 = ax.bar(index + bar_width/2, runs_by_over_team['United States of America'], bar_width, label='USA')

ax.set_xlabel('Over')
ax.set_ylabel('Runs')
ax.set_title('Total Runs in Each Over IND Vs USA T20')
ax.set_xticks(index)
ax.legend()
plt.show()
```

```python
# Convert categorical to numeric

wickets_lost = df.groupby(['over', 'team'])['player_out_name'].nunique().reset_index()
wickets_lost.rename(columns={'player_out_name': 'wickets_lost'}, inplace=True)
```

```python
#Calculating wicket loss by each team in each over
wicket_by_over_team = wickets_lost.groupby(['over', 'team'])['wickets_lost'].sum().unstack()
#print(wicket_by_over_team)

#plotting bar plot
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = wicket_by_over_team.index

bar1 = ax.bar(index - bar_width/2, wicket_by_over_team['India'], bar_width, label='IND')
bar2 = ax.bar(index + bar_width/2, wicket_by_over_team['United States of America'], bar_width, label='USA')

ax.set_xlabel('Over')
ax.set_ylabel('Runs')
ax.set_title('Total Wicket in Each Over IND Vs USA T20')
ax.set_xticks(index)
ax.legend(loc="upper right")
plt.show()
```

```python
#Run rate change over the course of inning of each team
df["cummulative_run"] = df.groupby(["over", "team"])["runs_total"].cumsum()
df["cummulative_over"] = df.groupby(["over", "team"])["over"].cumsum()+1
df["run_rate"] = (df["cummulative_run"] / df["cummulative_over"]) * 100
#print("Run Rate change over the course of inning of each team")
#print(df[["over", "team", "run_rate"]])

#plotting line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data = df, x = "over", y = "run_rate",  hue='team', marker='o')
plt.title("Run Rate change over the course of inning of each team")
plt.xlabel("Over")
plt.ylabel("Total Run")
plt.legend(title='Team')
plt.tight_layout()
plt.show()
```
